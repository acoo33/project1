```python
import pandas as pd
import numpy as np
```

```python
# The path to our CSV file
file = "./Resources/open-beer-database.csv"

# Read our Kickstarter data into pandas
df = pd.read_csv(file, encoding= 'unicode_escape')
df.head()
```

```python
us_beer = df.loc[df['Country'] == 'United States']
us_beer
```

```python
us_beer = us_beer[['Name','id','Alcohol By Volume','International Bitterness Units','Standard Reference Method', 'Style',  
                 'Category', 'Brewer', 'City', 'State', 'Coordinates'
                 ]]
us_beer
```

```python
us_beer_df = us_beer.dropna(how = 'any')
```

```python
us_beer_df['Region'] = ''
```

```python
northeast = ['Maine', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New Hampshire', 'Vermont', 'New York', 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland']
southeast = ['West Virginia', 'Virginia', 'Kentucky', 'Tennessee', 'North Carolina', 'South Carolina', 'Georgia', 'Alabama', 'Mississippi', 'Arkansas', 'Louisiana', 'Florida']
midwest = ['Ohio', 'Indiana', 'Michigan', 'Illinois', 'Missouri', 'Wisconsin', 'Minnesota', 'Iowa', 'Kansas', 'Nebraska', 'South Dakota', 'North Dakota']
southwest = ['Texas', 'Oklahoma', 'New Mexico', 'Arizona']
west = ['Colorado', 'Wyoming', 'Montana', 'Idaho', 'Washington', 'Oregon', 'Utah', 'Nevada', 'California', 'Alaska', 'Hawaii']
```

```python
us_beer_df['Region'] = np.where(us_beer_df['State'].isin(northeast), 'Northeast',
                      np.where(us_beer_df['State'].isin(southeast), 'Southeast',
                      np.where(us_beer_df['State'].isin(midwest), 'Midwest',
                      np.where(us_beer_df['State'].isin(southwest), 'Southwest',
                      np.where(us_beer_df['State'].isin(west), 'West',
                      'Unknown')))))
```

```python
beer_df = us_beer_df.loc[us_beer_df['Alcohol By Volume'] != 0]
beer_df
```

```python
beer_df.to_csv("./output_data/craft_beer.csv", index=False, header=True)
```
