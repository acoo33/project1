```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = "../output_data/craft_beer_nv.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
new_df = df.groupby('Region').size()
new_df

print(new_df)


regions = ["Midwest", "Northeast", "Southeast", "Southwest", "Unknown", "West"]
plt.bar(regions,new_df, color=['yellow', 'blue', 'red', 'black', 'orange'])
plt.title("Breweries per Region")
plt.ylabel("Quantity")
plt.xlabel("Region")
```
