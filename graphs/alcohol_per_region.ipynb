```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
```

```python
path = "../output_data/craft_beer_nv.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
new_df = df.groupby('Region')['Alcohol By Volume'].mean().round(2)
new_df = new_df.drop('Unknown')
new_df.head()
```

```python
#Creating a bar chart for # of Breweries per Region
labels = new_df.index
brewery_count = new_df.array

x = np.arange(len(labels))  # the label locations
width = 0.60  # the width of the bars

fig, ax = plt.subplots(facecolor='xkcd:silver')
rects1 = ax.bar(x, brewery_count, width, color=['yellow', 'blue', 'red', 'black', 'orange'])

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.title("Alcohol % per Region")
plt.ylabel("Alcohol %")
plt.xlabel("Region")

ax.set_xticks(x)
ax.set_ylim([5.25, 7.00])
ax.set_xticklabels(labels)
ax.set_facecolor('xkcd:silver')


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)

plt.savefig("../Images/Alcohol_Percent_Region.png")

plt.show()

fig.tight_layout()

plt.show()
```
