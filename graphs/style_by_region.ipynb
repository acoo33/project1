```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = "output_data/craft_beer.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
x_axis = df.Style.unique()
new_x_axis = list(x_axis)
new_x_axis
```

```python
new_df = df.groupby('Style').size()
new_df
```

```python
plt.bar(new_x_axis,new_df, color="r", width=2)
plt.xticks(rotation=90)


# plt.title("Breweries per Region")
# plt.ylabel("Quantity")
# plt.xlabel("Region")
#stackoverflow - scottlittle
```
